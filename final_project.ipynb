## **Step 2: Data Pre-processing**

**A) Data Cleaning**

Data cleaning means fixing or removing anything in the dataset that could cause wrong conclusions. Real-world data often contains mistakes like missing values, extra spaces, spelling errors, duplicates, or weird numbers. During data cleaning, we check the dataset carefully and make sure everything is correct, consistent, and ready for analysis. In simple words, data cleaning makes your raw data neat and correct so your analysis becomes reliable.

**A1.Handling Missing Values and Duplicates**

Missing values happen when some rows don’t have information in certain columns. If we don’t handle them, it can cause wrong results or errors in our analysis. To solve this, we can either remove the rows with missing values or fill them with something meaningful like average, median, most common value, or zero. The goal is to make sure every column has proper values before we continue working with the data. Sometimes, the same row or entry appears more than once in the dataset. These repeated rows are called duplicates, and they can disturb your analysis — for example, they may increase totals or averages incorrectly. Removing duplicates means checking the dataset for repeated rows and deleting them so that each record appears only once. This helps keep your data clean and trustworthy.

Outliers are values that are extremely high or low compared to the rest of the data. These unusual values can happen because of typing mistakes, measurement errors, or rare events. If we don’t identify them, they might affect averages, trends, or machine learning models. Outlier detection helps us spot these unusual numbers and decide whether to keep them or remove them if they are errors. This improves the accuracy of our analysis.

**B1.Standardizing Price Column**

To analyse rental data properly, the price column must be converted into a single numeric format. In the raw dataset, the price values appear in different formats such as plain numbers, values written with words like “lac” or “crore”, and even with text labels like “BDT”. These mixed formats prevent proper calculations, comparisons, and outlier detection.

```python
import re
import numpy as np

def convert_price(x):
    if pd.isna(x):
        return np.nan

    x = str(x).lower().strip()

    nums = re.findall(r'\d+\.?\d*', x)
    if len(nums) == 0:
        return np.nan

    num = float(nums[0])

    # crore
    if 'crore' in x:
        return num * 10000000

    # lac / lakh
    if 'lac' in x or 'lakh' in x:
        return num * 100000

    # return number as default BDT
    return num

df['price_clean'] = df['price'].apply(convert_price)
```

```python
df['price_clean'] = df['price_clean'].fillna(df['price_clean'].median())
```

```python
import re
import numpy as np
import pandas as pd

def convert_price(x):
    if pd.isna(x):
        return np.nan

    x = str(x).lower().strip()

    nums = re.findall(r'\d+\.?\d*', x)


    if not nums:
        return np.nan

    num = float(nums[0])


    if 'crore' in x:
        return num * 10000000

    if 'lakh' in x or 'lac' in x:
        return num * 100000


    return num

df['price_cleaned'] = df['price'].apply(convert_price)


df['price_cleaned'] = df['price_cleaned'].fillna(df['price_cleaned'].median())
```

```python
def clean_floor(x):
    x = str(x).lower()


    if 'ground' in x:
        return 0

    nums = re.findall(r'\d+', x)
    if len(nums) > 0:
        return int(nums[0])
    return None

df['floor_cleaned'] = df['floor'].apply(clean_floor)
```

**Categorical columns**

```python
import re

def clean_numeric_value(x):
    if pd.isna(x):
        return None
    x = str(x)

    # Remove everything except digits and dots
    x = re.sub(r"[^0-9.]", "", x)

    # If multiple dots, keep only the first
    if x.count(".") > 1:
        parts = x.split(".")
        x = parts[0]  # take the part before first dot

    # Convert empty → None
    if x == "" or x == ".":
        return None

    return float(x)

# Apply to columns
df["property size"] = df["property size"].apply(clean_numeric_value)
df["price"] = df["price"].apply(clean_numeric_value)
```

```python
def categorize_property_size(x):
    if pd.isna(x):
        return None
    if x <= 2000:
        return "Small"
    elif x <= 3000:
        return "Medium"
    else:
        return "Luxury"

df["Property_Size_Category"] = df["property size"].apply(categorize_property_size)
```

```python
def categorize_price(x):
    if pd.isna(x):
        return None
    if x <= 100000:
        return "below 1 lakh"
    elif x <= 200000:
        return "1 lakh to 2 lakh"
    elif x <= 400000:
        return "2 lakh to 4 lakh"
    else:
        return "above 4 lakh"

df["Price_Category"] = df["price"].apply(categorize_price)

```

```python
def categorize_year(x):
    if pd.isna(x): return None
    x = int(x)
    if x < 2000: return "Old"
    elif x <= 2015: return "Mid"
    return "New"

df["Year_Category"] = df["year built"].apply(categorize_year)
```

```python
import re

def clean_floor(x):
    if pd.isna(x):
        return None

    x = str(x).lower().strip()


    if "ground" in x or "g" == x:
        return 0
    if "basement" in x or "b" == x:
        return -1


    nums = re.findall(r"\d+", x)
    if len(nums) > 0:
        return int(nums[0])

    return None

df["floor_cleaned"] = df["floor"].apply(clean_floor)
```

```python
def categorize_floor(x):
    if pd.isna(x):
        return "basement"
    x = int(x)

    if x == -1:
        return "ground floor"
    elif x == 0:
        return "first Floor"
    elif 1 <= x <= 3:
        return "second Floor"
    elif 4 <= x <= 8:
        return "third Floor"
    elif 9 <= x <= 20:
        return "fourth Floor"
    else:
        return "fifth floor"

df["Floor_Category"] = df["floor_cleaned"].apply(categorize_floor)
```

## **Step 3 : Exploratory Data Analysis & Visualizations**

Exploratory Data Analysis (EDA) is a critical step in understanding the structure, patterns, and relationships within a dataset. After preprocessing, EDA helps uncover meaningful insights using statistical summaries, comparisons, and visualizations. This section combines both descriptive analysis and graphical exploration



Importing Required Visualization Libraries
To create plots and visualizations in Python, we use two main libraries:

● Matplotlib: The foundational plotting library in Python. Used for basic charts like line plots, bar charts, scatter plots, etc.

● Seaborn: A powerful visualization library built on top of Matplotlib. Provides advanced statistical plots with clean visuals.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('rental_listings.csv')

print("Dataset shape:", df.shape)
print(df.info())
print(df.describe(include='all').T)

plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap='viridis')
plt.title("Missing Values Heatmap")
plt.show()
```

**Univariate Analysis**



Univariate analysis focuses on examining one variable at a time. This helps us understand the individual characteristics of each feature in the dataset.

**Distribution of Price**

Understanding the distribution of price values helps identify how price range vary across all properties. A histogram is a useful tool for visualizing the frequency of different price ranges, while the KDE (Kernel Density Estimate) curve provides a smoothed representation of the distribution.

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Price_Category'], kde=True, bins=30)
plt.title("Distribution of Price")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

**Boxplot**

It helps you detect outliers in the price data.

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['Price_Category'])
plt.title("Boxplot of Cleaned Price")
plt.xlabel("Price")
plt.show()
```

**Histogram of property size**

Analyzing the distribution of property size helps us understand the actual size of the property.A histogram helps visualize the number of items falling into specific size.

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Property_Size_Category'], kde=True, bins=30, color='green')
plt.title("Distribution of Property Size")
plt.xlabel("Property Size")
plt.ylabel("Count")
plt.show()
```

**Floor Column**

Understanding the distribution of floor column helps to identify the floor which the property belongs.

```python
plt.figure(figsize=(10, 5))
sns.countplot(x=df['Floor_Category'])
plt.title("Count of Listings by Floor (Cleaned)")
plt.xlabel("Floor (numeric, 0 = Ground)")
plt.ylabel("Number of Listings")
plt.show()
```

**Bivariate Analysis**

Bivariate analysis studies the relationship between two variables at a time. It reveals how one feature influences or interacts with another.


**Scatter plot** : Property size vs Price

**Boxplot: Price by Floor**

**Multivariate Analysis**

Multivariate analysis examines interactions among three or more variables simultaneously. It reveals complex patterns not visible in univariate or bivariate analysis

**Barplot** : average price by floor
